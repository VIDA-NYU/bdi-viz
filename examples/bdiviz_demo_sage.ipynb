#### Scenario:
The Clinical Proteomic Tumor Analysis Consortium (CPTAC) program wants to make the data from its multiple studies readily available in an integrated, standardized format in the Cancer Research  Data Commons (CRDC). Much of the patient case and sample data are available in supplemental tables in the original primary research papers, but with varying variable names and value types. This exercise is based on the study by Dou et al, Proteogenomic data and resources for pan-cancer analysis: Cancer Cell, which integrated and harmonized data from ten CPTAC source studies. 

#### Task:
Semi-automatically extract and integrate patient case metadata from the ten source studies. Create a harmonized dataset in the Genomics Data Commons (GDC) data format. Include only patients that have tumor samples with proteogenomic data and were not excluded in the source studies.




```python
import pandas as pd
import numpy as np

import panel as pn
import bdikit as bdi
from bdiviz import BDISchemaMatchingHeatMap
```

### Loading the data

```python
source = pd.read_csv("datasets/dou.csv")
target = "gdc"

source
```

### Visualize Top Candidates in BDI-Viz Heatmap


#### Spotlighted Features

- **Interactive Heatmap**: Click on a heatmap cell to select candidates.
- **Candidate Manipulation**:
  - Use the **Accept Match**, **Reject Match**, and **Discard Column** buttons to manage matching candidates.
  - Undo or redo actions with the **Undo** and **Redo** buttons.
- **Value Comparisons**: Explore similar candidate values in detail.
- **Detailed Analysis**: Examine value distributions and schema descriptions for selected attributes.
- **Filtering Options**:
  - Filter by candidate data type using the **Candidate Type Selector**.
  - Identify similar source columns with the **Similar Sources Slider** (based on embeddings).
  - Adjust threshold values with the **Candidate Threshold Slider** to refine matches.

```python
heatmap_manager = BDISchemaMatchingHeatMap(
    source,
    target=target,
    top_k=40,
    ai_assitant=False,
)

heatmap_manager.plot_heatmap()
```


### Pass Results to BDI-Kit

BDI-Viz seamlessly integrates as an extension of BDI-Kit. By passing results from BDI-Viz to BDI-Kit, you can retain and utilize all manually updated candidates for further processing. 

```python
from bdikit.mapping_algorithms.column_mapping.algorithms import TwoPhaseSchemaMatcher

two_phase_viz = TwoPhaseSchemaMatcher(top_k_matcher=heatmap_manager)
column_mappings = bdi.match_schema(source, target=target, method=two_phase_viz)
column_mappings
```

```python
column_mappings = column_mappings[column_mappings["target"].str.strip().astype(bool)]
mappings = bdi.match_values(
    source,
    column_mapping=column_mappings,
    target=target,
    method="tfidf",
)

for mapping in mappings:
    print(f"{mapping.attrs['source']} => {mapping.attrs['target']}")
    display(mapping)
    print("")
```
